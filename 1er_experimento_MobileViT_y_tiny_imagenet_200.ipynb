```python
#1. Carga de datos
import os
import torch
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
from torch.utils.data import random_split

# 1. Transformaciones para Tiny-ImageNet
transform = transforms.Compose([
    transforms.Resize((224, 224)),  # Cambiar el tamaño a 224x224 requerido por MobileViT
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])  # Normalización estándar para RGB
])

# 2. Ruta del dataset
dataset_path = "/content/drive/MyDrive/Colab_Notebooks/dataset/images/tiny-imagenet-200/tiny-imagenet-200-64x64"

# 3. Cargar el conjunto de entrenamiento directamente
train_path = os.path.join(dataset_path, "train")
train_dataset = datasets.ImageFolder(root=train_path, transform=transform)

# 4. Reorganizar y cargar el conjunto de validación
val_path = os.path.join(dataset_path, "val")
val_annotations_path = os.path.join(val_path, "val_annotations.txt")

# Crear un mapeo de imágenes a clases para el conjunto de validación
val_labels = {}
with open(val_annotations_path, "r") as f:
    for line in f.readlines():
        parts = line.strip().split("\t")
        image_name, class_name = parts[0], parts[1]
        val_labels[image_name] = class_name

# Crear una carpeta temporal con subcarpetas organizadas por clase
temp_val_path = os.path.join(dataset_path, "temp_val")
os.makedirs(temp_val_path, exist_ok=True)

# Crear subcarpetas por clase y mover las imágenes
for image_name, class_name in val_labels.items():
    class_folder = os.path.join(temp_val_path, class_name)
    os.makedirs(class_folder, exist_ok=True)
    src_image_path = os.path.join(val_path, "images", image_name)
    dest_image_path = os.path.join(class_folder, image_name)
    if not os.path.exists(dest_image_path):  # Evitar duplicados si ya se procesó
        os.rename(src_image_path, dest_image_path)

# Cargar el conjunto de validación reorganizado
val_dataset = datasets.ImageFolder(root=temp_val_path, transform=transform)

# 5. Dividir el conjunto de entrenamiento si es necesario
train_size = int(0.8 * len(train_dataset))
val_size = len(train_dataset) - train_size
train_dataset, val_dataset = random_split(train_dataset, [train_size, val_size])

# 6. Crear DataLoaders
batch_size = 32
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=2)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=2)

print("Datasets y DataLoaders creados exitosamente.")
print(f"Tamaño del conjunto de entrenamiento: {len(train_loader.dataset)}")
print(f"Tamaño del conjunto de validación: {len(val_loader.dataset)}")

```

```python
# 2. Configurar el modelo MobileViT
# Importar el modelo MobileViT desde el repositorio
from qai_hub_models.models.mobile_vit.model import MobileVIT
import torch.nn as nn
import torch.optim as optim

# Cargar el modelo MobileViT preentrenado en ImageNet
model = MobileVIT.from_pretrained()

# Ajustar la última capa para 200 clases de Tiny-ImageNet
num_classes = 200
model.net.classifier = nn.Linear(model.net.classifier.in_features, num_classes)

# 3. Configurar el optimizador y la función de pérdida
criterion = nn.CrossEntropyLoss()  # Pérdida de clasificación estándar
# Es una elección estándar para optimizadores como Adam, balanceando convergencia rápida y estabilidad.
optimizer = optim.Adam(model.parameters(), lr=0.0001)  # Tasa de aprendizaje ajustada para fine-tuning
```

```python
# 4. Entrenar y evaluar el modelo en Tiny-ImageNet-200
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
import torch
from PIL import Image

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

# Inicializar listas para métricas
train_losses = []  # Pérdida promedio en cada época
val_accuracies = []  # Precisión promedio en validación
evaluated_images = []  # Para almacenar imágenes de validación
evaluated_predictions = []  # Para almacenar predicciones de validación

# Obtener nombres de las clases de Tiny-ImageNet
classes = train_loader.dataset.dataset.classes  # Clases desde ImageFolder

# Entrenamiento y evaluación
num_epochs = 10
for epoch in range(num_epochs):
    # Entrenamiento
    model.train()
    running_loss = 0.0
    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(inputs)  # Forward
        loss = criterion(outputs, labels)  # Calcular pérdida
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    train_losses.append(running_loss / len(train_loader))  # Pérdida promedio de la época

    # Evaluación en el conjunto de validación
    model.eval()
    all_labels = []
    all_predictions = []
    evaluated_images.clear()
    evaluated_predictions.clear()
    with torch.no_grad():
        for inputs, labels in val_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)  # Forward
            _, predictions = torch.max(outputs, 1)

            # Guardar etiquetas y predicciones
            all_labels.extend(labels.cpu().numpy())
            all_predictions.extend(predictions.cpu().numpy())

            # Guardar imágenes y predicciones para visualización
            if len(evaluated_images) < 16:  # Limitar a 16 imágenes
                evaluated_images.extend(inputs.cpu())
                evaluated_predictions.extend(predictions.cpu().numpy())

    # Calcular precisión
    accuracy = accuracy_score(all_labels, all_predictions)
    val_accuracies.append(accuracy * 100)  # Precisión en porcentaje
    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {train_losses[-1]:.4f}, Validation Accuracy: {val_accuracies[-1]:.2f}%")

# --- Graficar métricas ---
plt.figure(figsize=(12, 5))

# Pérdida de entrenamiento
plt.subplot(1, 2, 1)
plt.plot(range(1, len(train_losses) + 1), train_losses, label="Train Loss", marker='o')
plt.xlabel("Épocas")
plt.ylabel("Pérdida")
plt.title("Pérdida durante el entrenamiento")
plt.legend()

# Precisión en validación
plt.subplot(1, 2, 2)
plt.plot(range(1, len(val_accuracies) + 1), val_accuracies, label="Validation Accuracy", color="orange", marker='o')
plt.xlabel("Épocas")
plt.ylabel("Precisión (%)")
plt.title("Precisión en validación")
plt.legend()

plt.tight_layout()
plt.show()

# --- Cargar mapeo de códigos a descripciones desde words.txt ---
def load_class_dict(words_file_path):
    """
    Carga el diccionario de clases desde un archivo words.txt.

    Args:
        words_file_path: Ruta al archivo words.txt.

    Returns:
        dict: Diccionario que mapea códigos alfanuméricos a descripciones conceptuales.
    """
    class_dict = {}
    with open(words_file_path, "r") as f:
        for line in f:
            code, description = line.strip().split("\t", 1)
            class_dict[code] = description
    return class_dict

# Ruta al archivo words.txt
words_file_path = "/content/drive/MyDrive/Colab_Notebooks/dataset/tiny-imagenet-200-64x64/words.txt"
class_dict = load_class_dict(words_file_path)

# --- Mostrar imágenes evaluadas ---
def show_image_grid(images, predictions, class_dict, rows=4, cols=4):
    """
    Muestra un grid de imágenes con sus predicciones utilizando nombres de clases descriptivos.

    Args:
        images: Lista de imágenes (tensores).
        predictions: Lista de predicciones.
        class_dict: Diccionario que mapea códigos alfanuméricos a descripciones.
        rows: Número de filas en el grid.
        cols: Número de columnas en el grid.
    """
    fig, axes = plt.subplots(rows, cols, figsize=(15, 30))
    axes = axes.flatten()

    for img, pred, ax in zip(images[:rows * cols], predictions[:rows * cols], axes):
        # Convertir tensor a imagen y desnormalizar
        img = img.permute(1, 2, 0).numpy()
        img = (img * 0.5) + 0.5  # Desnormalizar
        img = img.clip(0, 1)  # Asegurar rango [0, 1]

        # Obtener la descripción conceptual de la predicción
        class_name = class_dict.get(pred, "Unknown")
        ax.imshow(img)
        ax.set_title(f"Pred: {class_name}", fontsize=10)
        ax.axis('off')

    for ax in axes[len(images):]:
        ax.axis('off')

    plt.tight_layout()
    plt.show()

# Mostrar imágenes con descripciones en lugar de códigos
show_image_grid(evaluated_images, evaluated_predictions, class_dict, rows=4, cols=4)


# 6. Guardar el modelo entrenado
#    Guarda el modelo fine-tuneado para su uso posterior.
%cd /content/drive/MyDrive/Colab_Notebooks/build/mobile_vit
torch.save(model.state_dict(), "Fine_tune_Mobile_ViT_tiny-imagenet-200_v2.pth")
import os
print(f"Modelo salvado en la ruta: {os.getcwd()}")
!ls
```
